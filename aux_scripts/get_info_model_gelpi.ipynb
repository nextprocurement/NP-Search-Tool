```python
info_docs = {}
for id_int, df_row in rows_without_nans.iterrows():
    id_p = df_row['place_id']
    this_info_doc = {}
    for k in range(len(tpc_labels)):
        if thetas[id_int, k] > 0:
            this_info_doc[k] = str(thetas[id_int, k])
    info_docs[id_p] = this_info_doc

len(info_docs)
```

```python
json_all = {
    "info_topics": info_tpcs,
    "info_docs": info_docs
}
with open('info_model_cpv45_11_tpcs.json', 'w') as json_file:
    json.dump(json_all, json_file, indent=4, ensure_ascii=False)
```

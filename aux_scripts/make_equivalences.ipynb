```python
from langdetect import detect
import re
```

```python
path_eq = "/export/usuarios_ml4ds/lbartolome/NextProcurement/NP-Search-Tool/sample_data/eq.txt"
```

```python
def det(x: str) -> str:
    """
    Detects the language of a given text

    Parameters
    ----------
    x : str
        Text whose language is to be detected

    Returns
    -------
    lang : str
        Language of the text
    """

    try:
        lang = detect(x)
    except:
        lang = 'Other'
    return lang
```

```python
pares_diccionario = {}
with open(path_eq, 'r') as archivo:
    for linea in archivo:
        linea = linea.strip()
        palabras = linea.split(':')
        patron = r'\b{}\b'.format(re.escape(palabras[0]))
        pares_diccionario[patron] = palabras[1]
pares_diccionario = dict(sorted(pares_diccionario.items(), key=lambda x: x[0]))
```

```python
sentence = "cpp_cvmc cvmc"
```

```python
def reemplazar_palabras(texto, diccionario):
    for palabra_original, palabra_nueva in diccionario.items():
        texto = re.sub(palabra_original, palabra_nueva, texto)
    return texto
reemplazar_palabras(sentence, pares_diccionario)
```
